## Task 1: Conceptual Understanding

---

### 1. Difference between "Love" and "love" in NLP

In Natural Language Processing, "Love" and "love" are treated as different tokens as the system is a case-sensitive system.This increases the vocabulary size unnecessarily as duplicates are present ,causes inconsistency in data.This affects the model performance.To handle this the text is often converted to lowercase.

---

### 2. What happens if stopwords are not removed?

If stopwords are not removed,the dataset contains unnecessary words like *is, the, and*,thus adds noise to the data.this  adds meaningless info and increases data size,and reduces accuracy.

---

### 3. When can removing stopwords be harmful?

Removing stopwords can sometimes change the meaning of a sentence when they provide some direction 
- **Sentiment Analysis:**  "I am not sad" → removing "not" changes meaning completely as the word not gives the negation in this sentence. 
- **Question Answering:**  "What is your name?" → removing "what" removes intent ,as it was directing towards the intent which was knowing the name.

---

### 4. Difference between Stemming and Lemmatization

Stemming :
- Method: Rule-based
- Output:  May not be a real word
- Accuracy: Lower

Lemmatization:
- Method: Dictionary-based
- Output: Always meaningful word
- Accuracy: Higher

## Task 2: Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    try:
        if not text or not isinstance(text, str):
            return [], ""

        #lowercase
        text = text.lower()

        #removing urls & emails
        text = re.sub(r'http\S+|www\S+|https\S+', '', text)
        text = re.sub(r'\S+@\S+', '', text)

        #removing numbers
        text = re.sub(r'\d+', '', text)

        #handling repeated characters (soooo → so)
        text = re.sub(r'(.)\1{2,}', r'\1', text)

        #removing special characters & emojis
        text = re.sub(r'[^a-z\s]', '', text)

        #removing extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        #Tokenization
        tokens = text.split()

        #Remove short words (except "no", "not")
        filtered_tokens = [
            word for word in tokens 
            if len(word) > 2 or word in ['no', 'not']
        ]

        cleaned_sentence = " ".join(filtered_tokens)

        return filtered_tokens, cleaned_sentence
    
    except Exception as e:
            print("Error processing text:", e)
            return [], ""
```

## Task 3: Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
```

## Task 4: Token Analytics

```python
def token_analysis(tokens):
    if not tokens:
        return 0, 0, 0
    
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total

    return total, unique, avg_len

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    total, unique, avg = token_analysis(tokens)

    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Total:", total, "Unique:", unique, "Avg Length:", avg)
```

## Task 5: Frequency Analysis

```python
from collections import Counter
all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent:", counter.most_common(10))

print("Least frequent:", counter.most_common()[-5:])
```

## Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)
print(result)
for i, sent in enumerate(result["clean_sentences"], 1):
    print(f"{i}. {sent}")
```
